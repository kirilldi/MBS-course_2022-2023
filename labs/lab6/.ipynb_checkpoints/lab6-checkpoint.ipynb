```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats
import math
from numpy.linalg import norm
from numpy import arccos, dot, pi, cross 
```

```python
def plot_circle(x,y,r): 
    angles=np.linspace(0,2*np.pi,50)
    x_cir=x+r*np.cos(angles)
    y_cir=y+r*np.sin(angles)
    plt.plot(x_cir,y_cir,"red")
```

```python
def plot_zone(x,y,distance,angle,fixang): #отрисовка интерферирующих объектов
    x1=x+distance*np.cos(angle)
    y1=y+distance*np.sin(angle)
    x2=x+distance*np.cos(angle+fixang)
    y2=y+distance*np.sin(angle+fixang)
    plt.plot([x,x1],[y,y1],"-.b")
    plt.plot([x,x2],[y,y2],"-.b")
    
```

```python
def poisson_point_process(lyam,R,area_size): #пуассоновское распределение точек внутри круга
    N=np.random.poisson(lyam*area_size**2)
    x1=[]
    y1=[]
    while len(x1)!=N:
        x=np.random.uniform(0,area_size)
        y=np.random.uniform(0,area_size)
        if((x-x0)**2+(y-y0)**2<=R*R):
            x1.append(x)
            y1.append(y)
    return x1, y1
       
```

```python
guard_size = 10 #размер зоны
guard_radius = 5 #радиус сигнала антенны
radius = 0.5 #радиус блокирующих устройств
distance = 5 #длина луча для отрисовки сектора

poisson_lam_block = 0.4
poisson_lam_int = 0.1

fixang=np.pi/4 #угол сектора интерферирующих устройств
list_point=np.array

x0=guard_size/2 #координаты расположения антенны
y0=guard_size/2
```

```python
x_block,y_block = poisson_point_process(poisson_lam_block,guard_radius,guard_size) #инициализируем блокирующие и интерферирующие устройства
x_int, y_int = poisson_point_process(poisson_lam_int,guard_radius,guard_size)
angle1=np.random.uniform(0,2*np.pi,size=len(x_int)) #массив углов поворота для интерфирирующих углов
```

```python
circle_angels=np.linspace(0,2*np.pi,100)
circle_x=x0+guard_radius*np.cos(circle_angels)
circle_y=y0+guard_radius*np.sin(circle_angels) #строим зону сигналов антенны
```

```python
plt.figure(dpi=100, figsize=(8,8),facecolor="olive")
plt.plot(x_block,y_block,".",alpha=0.7,label="fist",lw=5,mec="r",mew=3,ms=10)
for i in range(len(x_block)):
    plot_circle(x_block[i],y_block[i],radius)
plt.plot(x_int,y_int,'.',alpha=0.7,label="fist",lw=5,mec='b',mew=2,ms=10)
for i in range(len(x_int)):
    plot_zone(x_int[i],y_int[i],distance,angle1[i],fixang)
plt.plot(x0,y0,'.',alpha=0.7,label='fist',lw=5,mec='g',mew=2,ms=10)
plt.plot(circle_x,circle_y,'brown')
plt.xlim(0,guard_size)
plt.ylim(0,guard_size)
plt.show()
```

```python
def distance(x1,y1,x2,y2):
    return np.sqrt((x2-x1)**2+(y2-y1)**2)
```

```python
def check_pol(line_d1,line_d2,angle,fixang): #функция проверки нахождения Rx в зоне интерфирирующего передатчика
    x1,y1=line_d1
    x2,y2=line_d2
    atan1=math.atan2((y2-y1),(x2-x1))
    if atan1<0:
        atan1=2*np.pi+atan1
    d1=atan1>angle and atan1<(angle+fixang)
    if d1==True:
        return True
    else:
        return False
```

```python
def check_distance(A,B,C): #A,B - координаты приемника и передатчика, С - коордианаты центра блокатора
    CA=(C-A)/norm(C-A) #нормированный вектор из точки А в С
    BA=(B-A)/norm(B-A)
    CB=(C-B)/norm(C-B)
    AB=(A-B)/norm(A-B)
    #если BAC тупой
    if arccos(dot(CA,BA))>1: #скалярное произведение равно косинусу угла между векторами
        return norm(C-A)
    #если CBA тупой
    if arccos(dot(CB,AB))>1:
        return norm(C-B)
    return norm(cross(A-B,A-C))/norm(B-A)
```

```python
def crossing(x1,y1,x2,y2,x,y,radius_block):
    circle_point=[] #список координат всех центров окружности блокаторов 
    point_1=[]
    point_2=[]
    point_1.extend([x1,y1]) #координаты первой точки
    point_2.extend([x2,y2]) #координаты второй точки
    for i in range(len(x)):
        circle_point.append([x[i],y[i]])      
    for i in range(len(x)):
        if (np.round(check_distance(list_point(point_1),list_point(point_2),list_point(circle_point[i])),1)<=radius_block):
            return True
```

```python
def interf(x0,y0,x_int,y_int,P,fixang): 
    dis=distance(x0,y0,x_int,y_int)
    G=2/(1-np.cos(fixang/2))
    prx=P*G/(4*np.pi*dis**2)
    return prx
```

```python
P=2
num=1000
def imit(num,P,fixang,poisson_lam_block,poisson_lam_int,guard_radius,guard_size): #имитационное моделирование и подсчёт значений для графиков
    Interf=[]
    for j in range(num):
        Inter1=0
    x_block,y_block=poisson_point_process(poisson_lam_block,guard_radius,guard_size)
    x_int,y_int=poisson_point_process(poisson_lam_block,guard_radius,guard_size)
    angle1=np.random.uniform(0,2*np.pi,size=len(x_int))
    for i in range(len(x_int)):
        if check_pol([x_int[i],y_int[i]],[x0,y0],angle1[i],fixang)==True and crossing(x0,y0,x_int[i],y_int[i],x_block,y_block,radius)==None:
            Inter1=Inter1+interf(x0,y0,x_int[i],y_int[i],P,fixang)
    Interf.append(Inter1)
    return np.mean(Interf) #возвращаем среднее значение интерференции
```

```python
Interf_fi=[]
fixang=0
fixang_int=[]
step=10
for l in range(step):
    fixang=fixang+np.pi/step
    fixang_int.append(fixang)
    Interf_fi.append(imit(num,P,fixang,poisson_lam_block,poisson_lam_int,guard_radius,guard_size))

plt.figure(dpi=100)
plt.plot(fixang_int, Interf_fi)
plt.xlabel('Угол')
plt.ylabel('Уровень интерференции')
plt.show()
```

```python
Interf_fi=[]
fixang=np.pi/4
poisson_lam_int=0
poisson_lam_int_op=[]
for i in range(step):
    poisson_lam_int+=0.01
    poisson_lam_int_op.append(poisson_lam_int)
    Interf_fi.append(imit(num,P,fixang,poisson_lam_block,poisson_lam_int,guard_radius,guard_size))
    
plt.figure(dpi=100)
plt.plot(poisson_lam_int_op,Interf_fi)
plt.xlabel('Интенсивность поля интерферирующих устройств')    
plt.ylabel('Уровень интерференции')
plt.show()
```
