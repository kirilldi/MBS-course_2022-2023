## Исследование распределений и моментов связанных с преобразованием случайных величин


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gaussian_kde
```

Сгенерировать выборку точек, равномерно распределенных внутри круга двумя способами:
 
Равномерно распределить точки внутри квадрата, внутрь которого вписана окружность, и отфильтровать точки, лежащие за пределами окружности. Радиус окружности R=10, размер выборки 1000 точек.

```python
R = 10
set_size = 1000

set_x = []
set_y = []
dots_counter = 0

# Генерация точек
while dots_counter < set_size:
    x = np.random.uniform(-10,10)
    y = np.random.uniform(-10,10)
    
    if((x**2 + y**2) < R**2):
        set_x.append(x)
        set_y.append(y)
        
    dots_counter = dots_counter + 1
```

Для созданных выборок сделать следующее:
Создать рисунок, иллюстрирующий расположение точек сгенерированной выборки внутри окружности;

```python
# Окружность
angles = np.linspace(0,2*np.pi,100)
circle_x = R * np.cos(angles)
circle_y = R * np.sin(angles)
```

```python
plt.figure(figsize=(5,5))
plt.title("способ 1")
plt.plot(set_x,set_y,'.')
plt.plot(circle_x,circle_y)
```

```python
X = 20
Y = 0
```

```python
def get_distance(x1,x2,y1,y2):
    return np.sqrt((x2 - x1)**2 + (y2 - y1)**2)
```

```python
kde_distances_y = gaussian_kde(distances).evaluate(distances) #PDF
```

```python
plt.figure
plt.title("Плотность распределения расстояния от случайной точки в круге до точки [20,0]")
plt.plot(distances,kde_distances_y,color="r")
plt.hist(distances,bins=100,density=True,color="c");
```

Построить график плотности распределения расстояния между двумя случайными точками, равномерно расположенными внутри круга.

```python
# Переберу все комбинации точек
random_point_distances = []
set_x_copy = np.copy(set_x)
set_y_copy = np.copy(set_y)
i = 0 # counter
while(i<len(set_x)):
    for k in range(len(set_x)):
        random_point_distances.append(get_distance(set_x_copy[i],set_x_copy[k],set_y_copy[i],set_y_copy[k]))
    i += 1
    
```

```python
# Убираю все нули
for i in range(len(set_x_copy)):
    del(random_point_distances[random_point_distances.index(0.0)])
```
