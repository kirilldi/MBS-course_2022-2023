Сгенерировать три выборки размера 100, 1000 и 10000 для случайных расстояний между двумя точками, равномерно распределенные в прямоугольнике со сторонами 10 и 30.

```python
x_coords_100 = np.linspace(0,30,100)
y_coords_100 = np.linspace(0,10,100)
```

```python
def distance(x1,x2,y1,y2):
    return np.sqrt((x2 - x1)**2 + (y2 - y1)**2)
```

```python
def distance_line(number):
    a = 10
    b = 30
    distances = []
    for i in range(100):
        x1 = np.random.uniform(0,a)
        y1 = np.random.uniform(0,b)
        x2 = np.random.uniform(0,a)
        y2 = np.random.uniform(0,b)
        
        distances.append(distance(x1,x2,y1,y2))
        
    distances.sort()
    return distances
        
```

Получить среднее значение расстояния между точками

```python
average_dist_100 = np.average(distance_line(100))
average_dist_1000 = np.average(distance_line(1000))
average_dist_10000= np.average(distance_line(10000))
print(average_dist_100,average_dist_1000,average_dist_10000)
```
